```python
from sympy import Symbol, solve, Derivative

x = Symbol('x')
y = Symbol('y')

f = x**2 + y**2 + 3*x - 3*y + 4

d1 = Derivative(f, x).doit()
d2 = Derivative(f, y).doit()

critical_points_x = solve(d1, x)
critical_points_y = solve(d2, y)
s1 = Derivative(f, x, 2).doit()
s2 = Derivative(f, y, 2).doit()
s3 = Derivative(Derivative(f, y), x).doit()

delta = s1 * s2 - s3**2

results = []
for cx in critical_points_x:
    for cy in critical_points_y:
        delta_val = delta.subs({x: cx, y: cy}).evalf()
        s1_val = s1.subs({x: cx, y: cy}).evalf()
        f_val = f.subs({x: cx, y: cy}).evalf()
        if delta_val > 0 and s1_val < 0:
            nature = "Maximum"
        elif delta_val > 0 and s1_val > 0:
            nature = "Minimum"
        elif delta_val < 0:
            nature = "Saddle point"
        else:
            nature = "Indeterminate"
        results.append((cx, cy, f_val, delta_val, s1_val, nature))
results
```

```python
import numpy as np
from matplotlib import pyplot as plt
from sympy import *


x = Symbol('x')
y = sin(x) + cos(x)


x0 = float(0)
dy = diff(y, x)
d2y = diff(y, x, 2)
d3y = diff(y, x, 3)

yat = lambdify(x, y)
dyat = lambdify(x, dy)
d2yat = lambdify(x, d2y)
d3yat = lambdify(x, d3y)


y_taylor = yat(x0) + (x - x0) * dyat(x0) + ((x - x0)**2 / 2) * d2yat(x0) + ((x - x0)**3 / 6) * d3yat(x0)

print(simplify(y_taylor))

y_taylor_func = lambdify(x, y_taylor)
print("%.3f" % y_taylor_func(10 * (np.pi / 180)))


def f(x):
    return np.sin(x) + np.cos(x)


x_vals = np.linspace(-10, 10, 500)


plt.plot(x_vals, y_taylor_func(x_vals), color='red', label='Taylor Approximation')
plt.plot(x_vals, f(x_vals), color='green', label='Original Function')
plt.ylim([-3, 3])
plt.grid()
plt.legend()
plt.show()
```
